# Recreate the dataset


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [4, 4] # inches
mpl.rcParams['font.size'] = 12
# We set the random seed manually so that the results can be reproduced
np.random.seed(1234)
# define the data generating function
h = lambda x: np.sin(np.pi*x)
xmin = -1
xmax = 1
# Define a single independent variable
N = 101
x = np.linspace(xmin, xmax, num=N)
# Define the data with added noise
t = h(x) + 0.2*(np.random.randn(N))
# Add random Gaussian noise with unit variance
plt.figure()
plt.plot(x,h(x),'k-',x,t,'b.')
plt.xlabel('x')
plt.ylabel('t')
plt.tight_layout()
```

## Divide into test-validate-train

```python
# The training set
import matplotlib.pyplot as plt
xtrain = x[0::10]
ttrain = t[0::10]
# The validation set
xvalid = x[3::10]
tvalid = t[3::10]
# The test set
xtest = x[6::10]
ttest = t[6::10]

# Set the degree of polynomial fit
M = 11
# Empty list to record the RMS errors
rmstrain = []
rmsvalid = []
# Empty list to record the weights for each regression run
weights = []
plt.figure(figsize=[9,12])
# Loop over the regularisation parameter
loglambdavals = [-24,-18,-15,-12,-8,-6,-4,-2,-1,0,10]
for i, loglambda in enumerate(loglambdavals):
    # Create the Matrix of basis functions for the sampled data
    PhiTrain = np.array([pow(xtrain,j) for j in range(0,M)]).transpose()
    PhiValid = np.array([pow(xvalid,j) for j in range(0,M)]).transpose()
    PhiTest = np.array([pow(xtest,j) for j in range(0,M)]).transpose()
    # Create a high-resolution basis matrix so that we can plot the fitted polynomial smoothly
    hiPhi = np.array([pow(x,j) for j in range(0,M)]).transpose()
    # Solve the normal equations using a standard solver
    #print(np.matmul(PhiTrain.transpose(),PhiTrain))
    w = np.linalg.solve(np.matmul(PhiTrain.transpose(),PhiTrain)+np.exp(loglambda)*np.eye(ttrain.size), np.matmul(PhiTrain.transpose(),ttrain))
    # record the weights
    weights.append(w)
    # Compute the estimated values of the data points so we can compute the error
    ftrain = np.matmul(PhiTrain,w)
    fvalid = np.matmul(PhiValid,w)
    # compute the RMS error
    rmstrain.append(np.sqrt(np.sum(pow(ttrain-ftrain,2)/xtrain.size)))
    rmsvalid.append(np.sqrt(np.sum(pow(tvalid-fvalid,2)/xvalid.size)))
    # plot the data and the estimates of the model for the low-res and high-res samplings
    plt.subplot(4,3,i+1)
    plt.plot(x,h(x),'k-')
    plt.plot(xtrain,ttrain,'b.')
    plt.plot(xvalid,tvalid,'y.')
    plt.plot(xtest,ttest,'r.')
    plt.plot(x,np.matmul(hiPhi,w),'r--')
    plt.xlabel('x')
    plt.ylabel('t')
    plt.title(fr'$\ln\lambda={loglambda}$')
# Plot the error curve
plt.subplot(4,3,12)
plt.plot(loglambdavals,rmstrain,'b.-')
plt.plot(loglambdavals,rmsvalid,'y.-')
plt.xlabel(r'$\log\lambda$')
plt.ylabel('RMS Error')
plt.title('Training/Validation Loss')
plt.tight_layout()
plt.savefig('polyfit-l2')
# Print out the weights
for i,w in enumerate(weights):
    print(f"{loglambdavals[i]} & {' & '.join(f'{j:5.2f}' for j in w)} & {np.linalg.norm(w,ord=2):5.2f} \\\\")
```

```python
# Import Lasso
from sklearn import linear_model

# Set the degree of polynomial fit
M = 11
# Empty list to record the RMS errors
rmstrain = []
rmsvalid = []
# Empty list to record the weights for each regression run
weights = []
plt.figure(figsize=[9,12])
# Loop over the regularisation parameter
loglambdavals = [-24,-18,-15,-12,-8,-6,-4,-2,-1,0,10]
for i, loglambda in enumerate(loglambdavals):
    # Create the Matrix of basis functions for the sampled data
    PhiTrain = np.array([pow(xtrain,j) for j in range(0,M)]).transpose()
    PhiValid = np.array([pow(xvalid,j) for j in range(0,M)]).transpose()
    PhiTest = np.array([pow(xtest,j) for j in range(0,M)]).transpose()
    # Create a high-resolution basis matrix so that we can plot the fitted polynomial smoothly
    hiPhi = np.array([pow(x,j) for j in range(0,M)]).transpose()
    # Solve the normal equations using a standard solver
    #    w = np.linalg.solve(np.matmul(PhiTrain.transpose(),PhiTrain)+np.exp(loglambda)*np.eye(ttrain.size), np.matmul(PhiTrain.transpose(),ttrain))
    clf = linear_model.Lasso(alpha=np.exp(loglambda))
    clf.fit(PhiTrain,ttrain)
    # record the weights
    w = clf.coef_
    print(w)
    weights.append(w)
    # Compute the estimated values of the data points so we can compute the error
    ftrain = np.matmul(PhiTrain,w)
    fvalid = np.matmul(PhiValid,w)
    # compute the RMS error
    rmstrain.append(np.sqrt(np.sum(pow(ttrain-ftrain,2)/xtrain.size)))
    rmsvalid.append(np.sqrt(np.sum(pow(tvalid-fvalid,2)/xvalid.size)))
    # plot the data and the estimates of the model for the low-res and high-res samplings
    plt.subplot(4,3,i+1)
    plt.plot(x,h(x),'k-')
    plt.plot(xtrain,ttrain,'b.')
    plt.plot(xvalid,tvalid,'y.')
    plt.plot(xtest,ttest,'r.')
    plt.plot(x,np.matmul(hiPhi,w),'r--')
    plt.xlabel('x')
    plt.ylabel('t')
    plt.title(fr'$\ln\lambda={loglambda}$')
# Plot the error curve
plt.subplot(4,3,12)
plt.plot(loglambdavals,rmstrain,'b.-')
plt.plot(loglambdavals,rmsvalid,'y.-')
plt.xlabel(r'$\log\lambda$')
plt.ylabel('RMS Error')
plt.title('Training/Validation Loss')
plt.tight_layout()
plt.savefig('polyfit-l1')
# Print out the weights
for i,w in enumerate(weights):
    print(f"{loglambdavals[i]} & {' & '.join(f'{j:5.2f}' for j in w)} & {np.linalg.norm(w,ord=1):5.2f} \\\\")
```
